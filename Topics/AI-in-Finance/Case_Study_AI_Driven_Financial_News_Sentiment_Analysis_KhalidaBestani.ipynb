**5. Crawler Task (With Code)**

The crawler gathers headlines and article text from financial news websites such as Yahoo Finance.

```python
import requests
from bs4 import BeautifulSoup

def crawl_yahoo_news(stock):
    url = f"https://finance.yahoo.com/quote/{stock}/news"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    headlines = []
    for item in soup.find_all("h3"):
        headlines.append(item.get_text())

    return headlines
```

**6. Data Collection**

The crawler is executed daily for the selected stocks.

```python
stocks = ["MSFT", "GOOGL", "AAPL", "TSLA", "META"]
news_data = {}

for stock in stocks:
    news_data[stock] = crawl_yahoo_news(stock)
```

This creates a structured dataset of news headlines for each stock.

**7. AI Task: Sentiment Analysis Using Gen AI**
A Generative AI model (such as OpenAI GPT) is used to:

Classify sentiment as Positive, Negative, or Neutral

Identify keywords driving sentiment

```python
import openai
from google.colab import userdata

client = openai.OpenAI(api_key=userdata.get('OPENAI_API_KEY'))

def analyze_sentiment(news_text):
    prompt = f"""
    Analyze the sentiment of this financial news.
    Classify it as Positive, Negative, or Neutral.
    Identify keywords driving the sentiment.

    News: {news_text}
    """

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
```

**8. Daily Sentiment Processing**

Each stock’s news headlines are analyzed to determine overall daily sentiment.

```python
sentiment_results = {}

for stock, headlines in news_data.items():
    results = []
    for headline in headlines[:5]:  # limit for demo
        results.append(analyze_sentiment(headline))
    sentiment_results[stock] = results
```

**9. Insight Generation**

The Gen AI model summarizes sentiment into a single, human-readable insight per stock.

```python
def generate_insight(stock, sentiments):
    prompt = f"""
    Based on the sentiment results below,
    generate a one-line market insight.

    Stock: {stock}
    Sentiments: {sentiments}
    """

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
```
